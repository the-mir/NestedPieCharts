```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Marks = ([75,70],[70,80],[95,80],[70,75],[90,80])
Course = ("CSE","CE","EEE","ECE","ICE")
```

```python
fig =plt.figure(figsize=(3,4))
plt.pie(Marks,labels=Course,startangle=90, autopct='%1.1f%%',colors=['#510569','Yellow','Orange','Red','Green'],explode=[0.5,0,0,0,0])
plt.show()
```

```python
import matplotlib.pyplot as plt


outer_labels = ["CSE", "CE", "EEE", "ECE", "ICE"]
outer_sizes = [75, 70, 95, 70, 90]

inner_labels = ["CSE", "CE", "EEE", "ECE", "ICE"]
inner_sizes = [30, 20, 25, 35, 10]

fig, ax = plt.subplots()

ax.pie(outer_sizes, labels=outer_labels, startangle=90, autopct='%1.1f%%', colors=['#510569', 'Yellow', 'Orange', 'Red', 'Green'],
       wedgeprops=dict(width=0.4))


ax.pie(inner_sizes, labels=inner_labels, startangle=90, autopct='%1.1f%%',
       radius=0.6)

ax.axis('equal')

plt.show()
```
